```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/iris.csv')

df.head()
```

```python
df.groupby(['Class'])['Class'].count()
```

```python
dfg = df.groupby(['Class'])['Class'].count()
```

```python
dfg.plot.bar()
plt.title('Distribution of Iris Classes')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
```

```python
df['SepalLength'].plot.hist(bins=8, edgecolor='black')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
```

```python
df['SepalLength'].plot.box()
plt.title('Box Plot of Sepal Length')
plt.ylabel('Sepal Length (cm)')
```

```python
sns.violinplot(y=df['SepalLength'])
plt.title('Violin Plot of Sepal Length')
plt.ylabel('Sepal Length (cm)')
```
