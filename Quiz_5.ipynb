```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/iris.csv')

df.head()
```

```python
sns.scatterplot(data=df, x='SepalLength', y='SepalWidth', hue='Class', palette='bright')
plt.title('Sepal Length vs. Sepal Width by Class')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend(title='Class')
plt.show()
```
