```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/refs/heads/main/Data_Sets/Activity_Dataset_V1.csv')

df.head()
```

```python
plt.figure(figsize=(12, 8))
# Normalize 'calories' to a 0-1 range for alpha
normalized_calories = (df['calories'] - df['calories'].min()) / (df['calories'].max() - df['calories'].min())

# Call scatterplot without alpha, then set alpha on the collection manually
ax = sns.scatterplot(data=df, x='time', y='distance', hue='workout_type', size='max_heart_rate', style='vo2_max(%)', sizes=(20, 400))

# Apply the normalized calories as alpha to the collection of points
if ax.collections:
    ax.collections[0].set_alpha(normalized_calories)

plt.title('Distance vs. Time by Workout Type, Max Heart Rate, VO2 Max, and Calories')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
