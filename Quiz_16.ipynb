```python
data = {
    "age": ["20", "21", "19"],
    "gpa": ["3.5", "3.7", "3.9"],
    "enrolled": ["True", "False", "True"]
}

data["age"] = [int(x) for x in data["age"]]
data["gpa"] = [float(x) for x in data["gpa"]]
data["enrolled"] = [x == "True" for x in data["enrolled"]]

avg_gpa = sum(data["gpa"]) / len(data["gpa"])

print("Converted data:", data)
print("Average GPA:", avg_gpa)
```

## Incorrect Types
Age was set as string characters, and needed to be converted into int as they are whole numbers. GPA is also set as string characters, and needs to be converted to float for the decimal numbers. Enrollled needs to be boolean instead of string characters as it is a true false instead of characters.
