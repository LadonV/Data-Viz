```python
import re
import requests
import pandas as pa
from bs4 import BeautifulSoup
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
r = requests.get('https://en.wikipedia.org/wiki/List_of_highest_mountains_on_Earth', headers=headers)
html_contents = r.text
html_soup = BeautifulSoup(html_contents,"lxml")
tables = html_soup.find_all('table',class_="wikitable")

df1 = pa.read_html(str(tables))[0]
df1.head()
```

```python
df1.columns
```

```python
def clean_column(col):
    col = re.sub(r"\[.*?\]", "", col)
    col = re.sub(r"\(.*?\)", "", col)
    col = col.strip()
    col = col.lower()
    col = re.sub(r"\s+", "_", col)
    return col

df1.columns = [clean_column(col) for col in df1.columns]

print(df1.columns)
```

```python
df1['country_clean'] = df1['country/_region_administered_by'].str.replace(r"\\[.*?\\]", "", regex=True)
df1['country_list'] = df1['country_clean'].str.findall(r'[A-Z][a-z]+')

print(df1)
```
