```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/IOT-temp.csv')

df.head()
```

```python
df['noted_date'] = pd.to_datetime(df['noted_date'], format='%d-%m-%Y %H:%M')
df.head()
```

```python
min_date = df['noted_date'].min()
max_date = df['noted_date'].max()

print(f"Minimum Date and Time: {min_date}")
print(f"Maximum Date and Time: {max_date}")
```

```python
df_sorted = df.sort_values(by='noted_date')
df_sorted['time_diff'] = df_sorted['noted_date'].diff()

largest_interval = df_sorted['time_diff'].max()

print(f"The largest interval between temperature checks is: {largest_interval}")
```

```python
df['noted_date_timestamp'] = df['noted_date'].astype('int64') // 10**9

mean_timestamp = df['noted_date_timestamp'].mean()

average_datetime = pd.to_datetime(mean_timestamp, unit='s')

print(f"The average datetime is: {average_datetime}")
```

```python
specific_date = pd.to_datetime('09-11-2018', format='%d-%m-%Y')
outside_temp_on_date = df[(df['noted_date'].dt.date == specific_date.date()) & (df['out/in'] == 'Out')]

if not outside_temp_on_date.empty:
    average_outside_temp = outside_temp_on_date['temp'].mean()
    print(f"The average outside temperature on {specific_date.strftime('%d-%m-%Y')} was: {average_outside_temp:.2f} degrees")
else:
    print(f"No outside temperature data available for {specific_date.strftime('%d-%m-%Y')}")
```
