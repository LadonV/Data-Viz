```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/AB_NYC_2019.csv')

df.head()
```

```python
max_price_by_neighbourhood = df.groupby('neighbourhood_group')['price'].max().reset_index()

sns.barplot(x='neighbourhood_group', y='price', data=max_price_by_neighbourhood, hue='neighbourhood_group', palette='Set1')

plt.title('Maximum Price by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Maximum Price ($)')
plt.show()
```

```python
average_price_by_group_room = df.groupby(['neighbourhood_group', 'room_type'])['price'].mean().reset_index()

sns.barplot(x='neighbourhood_group', y='price', hue='room_type', data=average_price_by_group_room, palette='viridis')

plt.title('Average Price by Neighbourhood Group and Room Type')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price ($)')
plt.legend(title='Room Type')
plt.show()
```
