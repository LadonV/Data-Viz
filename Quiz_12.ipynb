```python
import requests
import pandas as pa
from bs4 import BeautifulSoup

# Define a User-Agent header to mimic a web browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

r = requests.get('https://en.wikipedia.org/wiki/Survivor_(American_TV_series)', headers=headers)
html_contents = r.text
html_soup = BeautifulSoup(html_contents,"lxml")
```

Gemini helped fix the code as the number of tables was reading as zero

```python
len(html_soup.find_all('table'))
```

```python
tables = html_soup.find_all('table',class_="wikitable")
tables[0].find_all('a')
```

```python
df = pa.read_html(str(tables))[0]
df
```

```python
# Export to CSV
df.to_csv("survivor_seasons.csv", index=False)
```
