```python
import pdfplumber
import pandas as pd

pdf_path = "sample-tables.pdf"

all_tables = []

with pdfplumber.open(pdf_path) as pdf:
    for page_num, page in enumerate(pdf.pages):
        print(f"Processing page {page_num + 1}")

        tables = page.extract_tables()

        for table in tables:
            df = pd.DataFrame(table)

            if not df.empty:
                proposed_columns = df.iloc[0].tolist()

                seen = {}
                unique_columns = []
                for col in proposed_columns:
                    col_str = str(col) if col is not None else ''
                    if col_str in seen:
                        seen[col_str] += 1
                        unique_columns.append(f"{col_str}_{seen[col_str]}")
                    else:
                        seen[col_str] = 0
                        unique_columns.append(col_str)

                df.columns = unique_columns
                df = df[1:].reset_index(drop=True)
                all_tables.append(df)
            else:
                print(f"Skipping empty table on page {page_num + 1}")

if all_tables:
    final_df = pd.concat(all_tables, ignore_index=True)

    final_df.to_csv("output.csv", index=False)

    print("Extraction complete! Saved as output.csv")
else:
    print("No tables extracted to save.")
```
