```python
url = "https://en.wikipedia.org/wiki/List_of_Survivor_(American_TV_series)_contestants"

# Add a User-Agent header to mimic a browser
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
response = requests.get(url, headers=headers)

# Use pandas.read_html to parse the content from the response
tables = pd.read_html(response.text)

winners_df = tables[0]  # usually first table
winners_df.head()
```

```python
winner_count = winners_df[winners_df['Finish'] == 'Winner'].shape[0]
print(f"Total count of 'Winner' in the 'Finish' column: {winner_count}")
```

```python
winner_rows_df = winners_df[winners_df['Finish'] == 'Winner']
print("DataFrame restricted to winner rows:")
print(winner_rows_df.head(10))
```
